```python
import numpy as np
import pandas as pd
from math import exp
```

```python
data = pd.read_csv(r"Social_Network_Ads.csv")
data.head()
```

```python
count_row = data.shape[0]
count_col = data.shape[1]
print(count_row,count_col)
```

```python
X=[]
X = data.Age
X
```

```python
Y = []
Y = data.Purchased
Y
```

```python
train_pct_index = int(0.2 * len(X))
X_train, X_test = X[:train_pct_index], X[train_pct_index:]
y_train, y_test = Y[:train_pct_index], Y[train_pct_index:]
```

```python
def normalize(X):
    return X - X.mean()


def predict(X, b0, b1):
    predict_array = []
    predict_array = np.array([1 / (1 + exp(-1*b0 + -1*b1*x)) for x in X])
    return predict_array


def logistic_regression(X, Y):

    X = normalize(X)

    b0 = 0
    b1 = 0
    L = 0.001  # learning rate
    epochs = 300

    for epoch in range(epochs):
        y_pred = predict(X, b0, b1)
        D_b0 = -2 * sum((Y - y_pred) * y_pred * (1 - y_pred))  
        D_b1 = -2 * sum(X * (Y - y_pred) * y_pred * (1 - y_pred))  
        b0 = b0 - L * D_b0
        b1 = b1 - L * D_b1
    
    return b0, b1
```

```python
b0, b1 = logistic_regression(X_train, y_train)

X_test_norm = normalize(X_test)
y_pred = predict(X_test_norm, b0, b1)
print(y_pred)
y_pred = [1 if p >= 0.5 else 0 for p in y_pred]
print(y_pred)
```

```python
accuracy = 0
for i in range(len(y_pred)):
    if y_test.iloc[i] == y_pred[i]:
        accuracy += 1
print(accuracy / float(len(y_test)) * 100.0)
```
